# FILTROVÁNÍ + ÚPRAVY

```python
# nový sloupec
df["new"] = df["age"] * 2

# přejmenování
df.rename(columns={"old": "new"}, inplace=True)

# odstranění sloupce
df.drop(columns=["col"], inplace=True)

# odstranění NaN
df.dropna(inplace=True)

# vyplnění NaN
df["age"].fillna(0, inplace=True)
```
